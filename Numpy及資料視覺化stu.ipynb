#使用plotly套件來畫3D圖

### 使用 Plotly 繪製 3D 曲面圖

Plotly 是一個功能強大的互動式圖形庫，非常適合繪製 3D 圖形。以下範例將繪製一個基於 `z = sin(x^2 + y^2)` 函數的 3D 曲面圖。

```
import plotly.graph_objects as go
import numpy as np

# 創建 x 和 y 數據
x=np.linspace(-10,10,200)
y=np.linspace(-10,10,200)
x, y = np.meshgrid(x, y)

# 計算 z 數據，這裡使用一個簡單的函數作為範例
z = x**2+4*x-6+y**2+6*y-4

# 創建 3D 表面圖
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# 設定圖形標題和軸標籤
fig.update_layout(
    title='3D 曲面圖範例',
    scene = dict(
        xaxis_title='X 軸',
        yaxis_title='Y 軸',
        zaxis_title='Z 軸'
    )
)

# 顯示圖形
fig.show()
```
